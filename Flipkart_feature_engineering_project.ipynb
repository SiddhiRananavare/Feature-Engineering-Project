### Data Cleaning

```python
df.isnull().sum() # no null values
```

```python
del df["Order ID"]
```

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
import datetime
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
df['day_of_week'] = df['Order Date'].dt.weekday
```

```python
df['is_weekend'] = df['day_of_week'].apply(lambda x:1 if x >= 5 else 0)
```

```python
df.head()
```

```python
# Total revenue check (if needed)
df['calculated_sales'] = df['Price (INR)'] * df['Quantity Sold']
```

```python
# Discount amount (actual loss due to discount)
df['discount_amount'] = (df['Price (INR)'] * df['Discount %']) / 100
```

```python
# Final price after discount
df['discounted_price'] = df['Price (INR)'] - df['discount_amount']
```

```python
# Profit per unit
df['profit_per_unit'] = df['Profit (INR)'] / df['Quantity Sold']
```

```python
# Quantity per order indicator
df['avg_order_value'] = df['Total Sales (INR)'] / df['Quantity Sold']
```

```python
# Value per rating (how much revenue per rating point)
df['value_per_rating'] = df['Total Sales (INR)'] / df['Customer Rating']
```

```python
df.head()
```

```python
# Previous sales
df['lag_1_sales'] = df['Total Sales (INR)'].shift(1)
df['lag_2_sales'] = df['Total Sales (INR)'].shift(2)
```

```python
# Previous profit
df['lag_1_profit'] = df['Profit (INR)'].shift(1)
```

```python
# Previous quantity
df['lag_1_quantity'] = df['Quantity Sold'].shift(1)
```

```python
df['rolling_mean_sales_3'] = df['Total Sales (INR)'].rolling(window=3).mean()
df['rolling_mean_sales_7'] = df['Total Sales (INR)'].rolling(window=7).mean()
```

```python
# Rolling sum
df['rolling_sum_sales_7'] = df['Total Sales (INR)'].rolling(window=7).sum()
```

```python
# Rolling profit trend
df['rolling_mean_profit_3'] = df['Profit (INR)'].rolling(3).mean()
```

```python
df.head(10)
```

### Encoding

```python
df = pd.get_dummies(df, columns=['Category','Region','Payment Method','Customer Segment'], drop_first=True)
```

```python
df.head()
```

### Feature Scaling

```python
num_cols = [
    'Price (INR)', 'Quantity Sold', 'Total Sales (INR)',
    'Profit (INR)', 'Discount %', 'Customer Rating'
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.head(10)
```

```python
import numpy as np
```

```python
df['price_log'] = np.log1p(df['Price (INR)'])
```

```python
df.head()
```

```python
Q1 = df['Price (INR)'].quantile(0.25)
Q3 = df['Price (INR)'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[(df['Price (INR)'] >= lower) & (df['Price (INR)'] <= upper)]
```
